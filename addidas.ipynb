##LÀM SẠCH DỮ LIỆU

```python
# BƯỚC 1: Loại bỏ các dòng dữ liệu bị trùng lặp hoàn toàn
df = df.drop_duplicates()
```

```python
df.info()
```

```python
# BƯỚC 2: Loại bỏ khoảng trắng dư thừa trong các cột dạng chuỗi
df['retailer'] = df['retailer'].str.strip()
df['state'] = df['state'].str.strip()
df['city'] = df['city'].str.strip()
```

```python
# BƯỚC 3: Chuẩn hoá định dạng chữ cho cột product để tránh phân nhóm sai
df['product'] = df['product'].str.title()
```

```python
# BƯỚC 4: Sửa lỗi chính tả trong cột region và đưa về NaN để xử lý sau
df['region'] = df['region'].replace('Unkown', np.nan)
```

```python
# BƯỚC 5: Chuyển cột invoice_date về kiểu datetime, giá trị lỗi sẽ thành NaT
df['invoice_date'] = pd.to_datetime(df['invoice_date'], errors='coerce')
```

```python
# BƯỚC 6: Làm sạch cột price_per_unit bằng cách loại bỏ dấu phẩy và ép kiểu float
df['price_per_unit'] = (
    df['price_per_unit']
    .astype(str)
    .str.replace(',', '', regex=False)
    .astype(float)
)
```

```python
# BƯỚC 7: Chuyển các cột số đang ở dạng object về numeric, lỗi chuyển thành NaN
df['total_sales'] = pd.to_numeric(df['total_sales'], errors='coerce')
df['operating_profit'] = pd.to_numeric(df['operating_profit'], errors='coerce')
```

```python
# BƯỚC 8: Phát hiện và loại bỏ outlier trong units_sold bằng phương pháp IQR
Q1 = df['units_sold'].quantile(0.25)          # Tính tứ phân vị thứ nhất
Q3 = df['units_sold'].quantile(0.75)          # Tính tứ phân vị thứ ba
IQR = Q3 - Q1                                 # Tính khoảng IQR
df = df[df['units_sold'] <= Q3 + 1.5 * IQR]   # Giữ lại các giá trị không phải outlier
```

```python
# BƯỚC 9: Chuẩn hoá dữ liệu rỗng trong cột sales_method
df['sales_method'] = df['sales_method'].replace('', np.nan)
```

```python
# BƯỚC 10: Điền missing values cho biến số bằng trung vị để giảm ảnh hưởng outlier
df['units_sold'] = df['units_sold'].fillna(df['units_sold'].median())
df['total_sales'] = df['total_sales'].fillna(df['total_sales'].median())
```

```python
# BƯỚC 11: Điền missing values cho biến phân loại bằng mode (giá trị phổ biến nhất)
df['region'] = df['region'].fillna(df['region'].mode()[0])
df['sales_method'] = df['sales_method'].fillna(df['sales_method'].mode()[0])
```

```python
# BƯỚC 12: Loại bỏ các dòng không có invoice_date vì không dùng được cho phân tích thời gian
df = df.dropna(subset=['invoice_date'])
```

Biểu đồ 1: Tổng doanh thu theo tháng (Line Chart)


```python
# Thiết lập template màu sắc bắt mắt
pio.templates.default = "plotly_white"
```

```python
# Chuyển đổi cột ngày tháng
df['invoice_date'] = pd.to_datetime(df['invoice_date'])

# Tạo cột năm, tháng, quý
df['year'] = df['invoice_date'].dt.year
df['month'] = df['invoice_date'].dt.month
df['quarter'] = df['invoice_date'].dt.quarter
df['year_month'] = df['invoice_date'].dt.strftime('%Y-%m')
```

```python
# Tính doanh thu theo tháng
monthly_revenue = df.groupby('year_month')['total_sales'].sum().reset_index()
# Nhóm theo tháng (year_month) và tính tổng doanh thu (total_sales)

monthly_revenue = monthly_revenue.sort_values('year_month')
# Sắp xếp theo tháng để đường biểu đồ chạy đúng thứ tự thời gian

fig1 = px.line(monthly_revenue, x='year_month', y='total_sales',
               title='📈 <b>Tổng Doanh Thu Theo Tháng</b>',
               labels={'total_sales': 'Doanh Thu (USD)', 'year_month': 'Tháng'},
               markers=True,
               line_shape='spline',
               color_discrete_sequence=['#FF6B6B'])
# Tạo biểu đồ đường:
# - Tiêu đề có icon và bold
# - Đổi tên nhãn trục
# - Hiển thị điểm marker
# - Đường cong mượt (spline)
# - Màu đỏ (#FF6B6B)

fig1.update_layout(
    xaxis_title="Tháng",
    yaxis_title="Doanh Thu (USD)",
    hovermode="x unified",  # Hiển thị thông tin tooltip theo trục x
    template="plotly_white",  # Nền trắng sạch sẽ
    height=500  # Chiều cao 500px
)
# Cập nhật bố cục biểu đồ

fig1.show()
# Hiển thị biểu đồ
```

Biểu đồ 2: Doanh thu theo vùng (Bar Chart)


```python
region_revenue = df.groupby('region')['total_sales'].sum().reset_index().sort_values('total_sales', ascending=False)
# Nhóm theo vùng, tính tổng doanh thu, sắp xếp giảm dần

fig2 = px.bar(region_revenue, x='region', y='total_sales',
              title='🗺️ <b>Doanh Thu Phân Bổ Theo Vùng</b>',
              labels={'total_sales': 'Doanh Thu (USD)', 'region': 'Vùng'},
              color='total_sales',  # Màu sắc theo giá trị doanh thu
              color_continuous_scale='Viridis',  # Dùng palette Viridis (xanh-lục)
              text=region_revenue['total_sales'].apply(lambda x: f'${x/1000000:.1f}M'))
# Tạo biểu đồ cột:
# - Tiêu đề có icon map và bold
# - Đổi tên nhãn trục
# - Màu gradient theo doanh thu
# - Hiển thị text trên cột (định dạng triệu USD)

fig2.update_layout(
    xaxis_title="Vùng",
    yaxis_title="Doanh Thu (USD)",
    showlegend=False,  # Ẩn chú thích (legend)
    height=500  # Chiều cao 500px
)
# Cập nhật bố cục biểu đồ

fig2.update_traces(textposition='outside')
# Đặt vị trí text ngoài đầu cột cho dễ đọc

fig2.show()
# Hiển thị biểu đồ
```

Biểu đồ 3: Doanh thu theo sản phẩm (Treemap)


```python
product_revenue = df.groupby('product')['total_sales'].sum().reset_index().sort_values('total_sales', ascending=False)
# Nhóm theo sản phẩm, tính tổng doanh thu, sắp xếp giảm dần

fig3 = px.treemap(product_revenue,
                  path=['product'],  # Phân cấp theo sản phẩm (chỉ 1 cấp)
                  values='total_sales',  # Kích thước ô theo doanh thu
                  title='👟 <b>Phân Bổ Doanh Thu Theo Dòng Sản Phẩm</b>',
                  color='total_sales',  # Màu sắc theo doanh thu
                  color_continuous_scale='RdBu',  # Dùng palette đỏ-xanh dương (Red-Blue)
                  hover_data={'total_sales': ':$,.0f'})  # Định dạng tooltip hiển thị doanh thu

fig3.update_layout(
    height=600,  # Chiều cao 600px
    margin=dict(t=50, l=25, r=25, b=25)  # Canh lề: top 50px, các cạnh 25px
)

fig3.show()
```

Biểu đồ 4: Doanh thu theo phương thức bán hàng (Pie Chart với Donut)


```python
method_revenue = df.groupby('sales_method')['total_sales'].sum().reset_index()
# Nhóm theo phương thức bán hàng, tính tổng doanh thu

fig4 = px.pie(method_revenue,
              values='total_sales',  # Giá trị dùng để tính tỷ lệ
              names='sales_method',  # Tên các phần (phương thức bán hàng)
              title='🛍️ <b>Tỷ Lệ Doanh Thu Theo Phương Thức Bán Hàng</b>',
              hole=0.4,  # Tạo donut chart với lỗ giữa 40%
              color_discrete_sequence=px.colors.qualitative.Set3)  # Palette màu phân loại

fig4.update_traces(textposition='inside',  # Đặt text bên trong các phần
                   textinfo='percent+label',  # Hiển thị % + tên phương thức
                   hovertemplate='<b>%{label}</b><br>Doanh thu: $%{value:,.0f}<br>Tỷ lệ: %{percent}')
# Template khi hover: tên + doanh thu + tỷ lệ

fig4.update_layout(height=500)  # Đặt chiều cao 500px

fig4.show()
```

Biểu đồ 5: Top 10 nhà bán lẻ có doanh thu cao nhất (Horizontal Bar)


```python
retailer_revenue = df.groupby('retailer')['total_sales'].sum().reset_index()
retailer_revenue = retailer_revenue.sort_values('total_sales', ascending=True).tail(10)
# Nhóm theo nhà bán lẻ, tính tổng doanh thu, lấy top 10 (sắp xếp tăng dần rồi lấy 10 cuối)

fig5 = px.bar(retailer_revenue,
               y='retailer',  # Trục y: tên nhà bán lẻ (để tạo bar ngang)
               x='total_sales',  # Trục x: doanh thu
               title='🏪 <b>Top 10 Nhà Bán Lẻ Có Doanh Thu Cao Nhất</b>',
               orientation='h',  # Biểu đồ ngang (horizontal)
               color='total_sales',  # Màu gradient theo doanh thu
               color_continuous_scale='Plasma',  # Dùng palette Plasma
               text=retailer_revenue['total_sales'].apply(lambda x: f'${x/1000000:.2f}M'))
# Định dạng text: hiển thị doanh thu theo đơn vị triệu USD (2 số thập phân)

fig5.update_layout(
    yaxis_title="Nhà bán lẻ",
    xaxis_title="Doanh Thu (USD)",
    showlegend=False,  # Ẩn chú thích
    height=500  # Chiều cao 500px
)

fig5.update_traces(textposition='outside')  # Đặt text bên ngoài cột

fig5.show()
```

Biểu đồ 6: Phân tích lợi nhuận biên theo sản phẩm (Bubble Chart)


```python
# Tính profit margin
product_profit = df.groupby('product').agg({
    'total_sales': 'sum',
    'operating_profit': 'sum',
    'units_sold': 'sum'
}).reset_index()

product_profit['profit_margin'] = (product_profit['operating_profit'] / product_profit['total_sales']) * 100
product_profit['avg_price'] = product_profit['total_sales'] / product_profit['units_sold']

fig6 = px.scatter(product_profit,
                  x='total_sales',
                  y='profit_margin',
                  size='units_sold',
                  color='product',
                  title='💹 <b>Phân Tích Lợi Nhuận Biên Theo Sản Phẩm</b>',
                  labels={'total_sales': 'Tổng Doanh Thu (USD)',
                          'profit_margin': 'Lợi Nhuận Biên (%)',
                          'units_sold': 'Số Lượng Bán',
                          'product': 'Sản Phẩm'},
                  hover_name='product',
                  hover_data={'avg_price': ':.2f'},
                  size_max=60)

fig6.update_layout(
    height=600,
    xaxis=dict(type='log'),
    showlegend=True
)
fig6.show()
```

Biểu đồ 7: Xu hướng doanh thu theo quý (Area Chart)


```python
quarterly_revenue = df.groupby(['year', 'quarter'])['total_sales'].sum().reset_index()
quarterly_revenue['period'] = quarterly_revenue['year'].astype(str) + '-Q' + quarterly_revenue['quarter'].astype(str)
# Nhóm theo năm và quý -> tính tổng doanh thu -> tạo cột 'period' (vd: 2023-Q1)

fig7 = px.area(quarterly_revenue,
               x='period',  # Trục x: quý (period)
               y='total_sales',  # Trục y: doanh thu
               title='📊 <b>Xu Hướng Doanh Thu Theo Quý</b>',
               labels={'total_sales': 'Doanh Thu (USD)', 'period': 'Quý'},
               markers=True,  # Hiển thị điểm marker
               color_discrete_sequence=['#00CC96'])  # Màu xanh lá (#00CC96)

fig7.update_layout(
    xaxis_title="Quý",
    yaxis_title="Doanh Thu (USD)",
    height=500  # Chiều cao 500px
)

fig7.update_traces(fill='tozeroy')  # Tô đầy vùng từ đường xuống trục y=0

fig7.show()
```

Biểu đồ 8: Heatmap doanh thu theo tháng và vùng


```python
# Tạo pivot table cho heatmap
heatmap_data = df.pivot_table(values='total_sales',
                               index='region',  # Hàng: vùng
                               columns='month',  # Cột: tháng
                               aggfunc='sum')  # Hàm tổng hợp: tính tổng

fig8 = go.Figure(data=go.Heatmap(
    z=heatmap_data.values,  # Ma trận giá trị doanh thu
    x=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],  # Nhãn tháng
    y=heatmap_data.index,  # Nhãn vùng
    colorscale='YlOrRd',  # Thang màu vàng-cam-đỏ (Yellow-Orange-Red)
    colorbar=dict(title="Doanh Thu (USD)")  # Thanh màu với tiêu đề
))

fig8.update_layout(
    title='🔥 <b>Heatmap Doanh Thu: Tháng vs Vùng</b>',  # Tiêu đề với icon lửa
    xaxis_title="Tháng",
    yaxis_title="Vùng",
    height=500  # Chiều cao 500px
)

fig8.show()
```

Biểu đồ 9: So sánh doanh thu Online vs Offline (Grouped Bar Chart)


```python
# Phân loại Online/Offline
df['sales_channel'] = df['sales_method'].apply(lambda x: 'Online' if x == 'Online' else 'Offline')
# Tạo cột mới 'sales_channel': Online nếu sales_method là 'Online', còn lại là Offline

channel_product_revenue = df.groupby(['sales_channel', 'product'])['total_sales'].sum().reset_index()
# Nhóm theo kênh bán và sản phẩm -> tính tổng doanh thu

fig9 = px.bar(channel_product_revenue,
              x='product',  # Trục x: sản phẩm
              y='total_sales',  # Trục y: doanh thu
              color='sales_channel',  # Màu phân biệt kênh bán (Online/Offline)
              barmode='group',  # Hiển thị cột nhóm cạnh nhau
              title='🔄 <b>So Sánh Doanh Thu Online vs Offline Theo Sản Phẩm</b>',
              labels={'total_sales': 'Doanh Thu (USD)', 'product': 'Sản Phẩm'},
              color_discrete_map={'Online': '#636EFA', 'Offline': '#EF553B'})  # Màu cố định cho mỗi kênh

fig9.update_layout(
    xaxis_title="Sản Phẩm",
    yaxis_title="Doanh Thu (USD)",
    height=600,  # Chiều cao 600px
    xaxis_tickangle=-45  # Xoay nhãn trục x 45 độ ngược chiều kim đồng hồ
)

fig9.show()
```

Biểu Đồ 10: Phân tích biến động doanh thu theo quý


```python
# Phân tích doanh thu theo quý và so sánh với quý trước
quarterly_revenue = df.groupby(['year', 'quarter'])['total_sales'].sum().reset_index()
quarterly_revenue['period'] = quarterly_revenue['year'].astype(str) + ' Q' + quarterly_revenue['quarter'].astype(str)

# Tính % thay đổi
quarterly_revenue['change'] = quarterly_revenue['total_sales'].pct_change() * 100
quarterly_revenue['change_type'] = quarterly_revenue['change'].apply(lambda x: 'increase' if x > 0 else 'decrease')

# Tạo waterfall chart
fig10_waterfall = go.Figure(go.Waterfall(
    name="Doanh Thu",
    orientation="v",
    measure=["absolute"] + ["relative"]*(len(quarterly_revenue)-2) + ["total"],
    x=quarterly_revenue['period'],
    y=quarterly_revenue['total_sales'],
    text=[f"${x/1000:.0f}K<br>{pct:+.1f}%" if i>0 else f"${x/1000:.0f}K"
          for i, (x, pct) in enumerate(zip(quarterly_revenue['total_sales'], quarterly_revenue['change']))],
    textposition="outside",
    connector={"line": {"color": "rgb(63, 63, 63)"}},
    increasing={"marker": {"color": "#00CC96"}},
    decreasing={"marker": {"color": "#EF553B"}},
    totals={"marker": {"color": "#636EFA"}}
))

fig10_waterfall.update_layout(
    title="🌊 <b>Phân Tích Biến Động Doanh Thu Theo Quý (Waterfall Chart)</b>",
    xaxis_title="Quý",
    yaxis_title="Doanh Thu (USD)",
    height=500,
    showlegend=False
)

fig10_waterfall.add_annotation(
    text="● Xanh: Tăng trưởng<br>● Đỏ: Sụt giảm<br>● Tím: Tổng cộng",
    xref="paper", yref="paper",
    x=0.02, y=0.95,
    showarrow=False,
    bgcolor="white",
    bordercolor="black",
    borderwidth=1
)

fig10_waterfall.show()
```

Biểu Đồ 11: Phân tích mối quan hệ giữa số lượng bán và doanh thu (Scatter Plot với Trendline)


```python
# Tính trung bình theo sản phẩm
product_analysis = df.groupby('product').agg({
    'units_sold': 'mean',
    'total_sales': 'mean',
    'price_per_unit': 'mean'
}).reset_index()

fig11 = px.scatter(product_analysis,
                   x='units_sold',
                   y='total_sales',
                   size='price_per_unit',
                   color='product',
                   title='📊 <b>Phân Tích Mối Quan Hệ: Số Lượng Bán vs Doanh Thu</b>',
                   labels={'units_sold': 'Số Lượng Bán Trung Bình',
                           'total_sales': 'Doanh Thu Trung Bình (USD)',
                           'price_per_unit': 'Giá Trung Bình',
                           'product': 'Sản Phẩm'},
                   hover_name='product',
                   trendline="ols",
                   trendline_scope="overall",
                   trendline_color_override="gray")

fig11.update_layout(
    xaxis_title="Số Lượng Bán Trung Bình",
    yaxis_title="Doanh Thu Trung Bình (USD)",
    height=600,
    showlegend=True
)
fig11.add_annotation(x=0.98, y=0.02,
                     text="Kích thước biểu thị giá trung bình",
                     showarrow=False,
                     xref="paper", yref="paper")
fig11.show()
```

Biểu Đồ 12: Phân tích doanh thu và lợi nhuận theo nhà bán lẻ (Dual Axis Bar Chart)


```python
# Top 8 nhà bán lẻ
top_retailers = df.groupby('retailer')['total_sales'].sum().nlargest(8).index.tolist()
retailer_df = df[df['retailer'].isin(top_retailers)]

retailer_summary = retailer_df.groupby('retailer').agg({
    'total_sales': 'sum',
    'operating_profit': 'sum'
}).reset_index()

retailer_summary = retailer_summary.sort_values('total_sales', ascending=True)

fig12 = go.Figure()

# Bar cho doanh thu
fig12.add_trace(go.Bar(
    x=retailer_summary['total_sales'],
    y=retailer_summary['retailer'],
    name='Doanh Thu',
    orientation='h',
    marker_color='#636EFA',
    hovertemplate='<b>%{y}</b><br>Doanh thu: $%{x:,.0f}<extra></extra>'
))

# Bar cho lợi nhuận (thứ cấp)
fig12.add_trace(go.Bar(
    x=retailer_summary['operating_profit'],
    y=retailer_summary['retailer'],
    name='Lợi Nhuận',
    orientation='h',
    marker_color='#00CC96',
    hovertemplate='<b>%{y}</b><br>Lợi nhuận: $%{x:,.0f}<extra></extra>'
))

fig12.update_layout(
    title='🏪 <b>Top 8 Nhà Bán Lẻ: So Sánh Doanh Thu vs Lợi Nhuận</b>',
    xaxis_title="Số Tiền (USD)",
    yaxis_title="Nhà Bán Lẻ",
    barmode='group',
    height=500,
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1)
)
fig12.show()
```

Biểu Đồ 13: Phân tích hiệu quả theo phương thức bán hàng (Radar Chart)


```python
# Tính các chỉ số hiệu suất theo phương thức bán hàng
method_metrics = df.groupby('sales_method').agg({
    'total_sales': 'sum',
    'operating_profit': 'sum',
    'units_sold': 'sum',
    'price_per_unit': 'mean'
}).reset_index()

# Tính thêm các chỉ số
method_metrics['avg_order_value'] = method_metrics['total_sales'] / method_metrics['units_sold']
method_metrics['profit_margin'] = (method_metrics['operating_profit'] / method_metrics['total_sales']) * 100

# Chuẩn hóa dữ liệu cho radar chart (0-100 scale)
from sklearn.preprocessing import MinMaxScaler

metrics_to_scale = ['total_sales', 'operating_profit', 'units_sold', 'avg_order_value', 'profit_margin']
scaler = MinMaxScaler()
method_metrics_scaled = method_metrics.copy()
method_metrics_scaled[metrics_to_scale] = scaler.fit_transform(method_metrics[metrics_to_scale]) * 100

# Tạo radar chart
categories = ['Doanh Thu', 'Lợi Nhuận', 'Số Lượng Bán', 'Giá Trị Đơn TB', 'Lợi Nhuận Biên (%)']

fig13 = go.Figure()

for i, row in method_metrics_scaled.iterrows():
    fig13.add_trace(go.Scatterpolar(
        r=[row['total_sales'], row['operating_profit'], row['units_sold'],
           row['avg_order_value'], row['profit_margin']],
        theta=categories,
        fill='toself',
        name=row['sales_method'],
        line_color=px.colors.qualitative.Set1[i]
    ))

fig13.update_layout(
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[0, 100]
        )),
    title='📊 <b>Radar Chart: Hiệu Quả Phương Thức Bán Hàng</b>',
    height=600,
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="center", x=0.5)
)
fig13.show()
```

Biểu Đồ 14: Phân tích theo mùa (Seasonal Analysis - Heatmap)


```python
# Tạo heatmap doanh thu theo tháng và năm
heatmap_data = df.groupby(['year', 'month'])['total_sales'].sum().reset_index()
heatmap_pivot = heatmap_data.pivot(index='month', columns='year', values='total_sales')

# Đổi tên tháng
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

fig14 = go.Figure(data=go.Heatmap(
    z=heatmap_pivot.values,
    x=heatmap_pivot.columns,
    y=month_names,
    colorscale='YlOrRd',
    colorbar=dict(title="Doanh Thu (USD)"),
    hoverongaps=False,
    text=heatmap_pivot.applymap(lambda x: f'${x/1000:.0f}K'),
    texttemplate="%{text}",
    textfont={"size": 10}
))

fig14.update_layout(
    title='📅 <b>Phân Tích Doanh Thu Theo Mùa Vụ</b>',
    xaxis_title="Năm",
    yaxis_title="Tháng",
    height=500
)
fig14.show()
```
